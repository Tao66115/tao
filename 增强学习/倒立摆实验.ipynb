```python
import gym
env = gym.make('CartPole-v0')      #增强学习 倒立摆训练
for i_episode in range(100):    
    observation = env.reset()   
    for t in range(100):           #训练一百次
        env.render()               #重置环境
        print(observation)         #输出 observation
        action = env.action_space.sample()     #设置action
        observation, reward, done, info = env.step(action)   
        if done:
            print("Episode finished after {} timesteps".format(t+1))   #如果倒立摆倒下，重置，返回obserbation
            break
env.close()                        #结束
```
